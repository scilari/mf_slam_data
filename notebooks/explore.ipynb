```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
files = [
    "../data/processed_csv/mapping_data/apartment_2011-03-27.csv",
    "../data/processed_csv/validation_data/cse_checkpoints_2011-03-27.csv"
]

for file in files[:]:
    df = pd.read_csv(file)
    
    spacebars = df[df.spacebar != 0] # Validation data spacebars (loop closure)
    df = df[::5] # Downsample for visualisation
        
    print(df.head())

    plt.figure()
    plt.scatter(df.x, df.y, edgecolors="black", facecolors="white", s=7.0)
    
    plt.scatter(spacebars.x, spacebars.y, edgecolors="black", facecolors="red", s=50.0)
    plt.axis("equal")
    plt.title(file)

    plt.figure()
    plt.plot(df.t, df.mx)
    plt.plot(df.t, df.my)
    plt.plot(df.t, df.mz)
    plt.title(file)
```
